The objective of this analysis is to uncover meaningful patterns and relationships within customer data that can enhance creditworthiness evaluation and support better business decision-making.

The dataset used in this study contains detailed customer-level information, including income, credit utilization, number of bank accounts, number of credit cards, outstanding debt, payment delays, and credit score classifications. The analysis began with a comprehensive data cleaning process, addressing missing values, correcting inconsistent data formats, and removing duplicate records to ensure data quality and reliability.

The exploratory analysis was structured to answer key business questions, such as whether higher income correlates with better credit scores, the impact of credit utilization on creditworthiness, and how payment behavior affects credit scores. The findings reveal that payment behavior is the most critical determinant of credit score. Customers with a higher number of delayed payments or longer delays from due dates consistently exhibit lower credit scores, highlighting the importance of timely repayments. Credit utilization ratio also shows a strong negative relationship with credit scores; individuals utilizing a large portion of their available credit are more likely to be classified as high-risk borrowers.

Income was found to have a positive but not definitive influence on credit scores. While higher-income individuals tend to have better scores, income alone does not guarantee creditworthiness, emphasizing the need to consider multiple factors simultaneously. Additionally, customers with a moderate number of credit cards and loans generally maintain healthier credit profiles, whereas excessive borrowing or credit usage indicates financial stress and increased risk.

The analysis also highlights the role of outstanding debt, where higher debt levels are associated with lower credit scores. Segmentation techniques were applied to categorize customers into risk groups based on their credit utilization, enabling clearer identification of low, medium, and high-risk individuals.

From a business perspective, these insights can significantly enhance Paisabazaar’s decision-making processes.
